## Задача проекта

На основе данных клиентов оператора сотовой связи проанализировать поведение клиентов и найти оптимальный тариф

## Описание проекта

Клиентам предлагают два тарифных плана: «Смарт» и «Ультра». Чтобы скорректировать рекламный бюджет, коммерческий департамент хочет понять, какой тариф приносит больше денег.

## Вывод

Тариф "Смарт" более популярен. Плата за дополнительные ресурсы значительно дороже чем в "Ультре", как результат доход "Смарт" приносит больше.

Не смотря на то, что в тарифе "Ультра" пользователей почти в два раз меньше и базовый пакет покрывает требования почти большинства, стабильная абонетская плата приносит хороший доход.

Наиболее выгоден для компании тариф "Смарт". Стабильная, не высокая абонентская плата, привлекает пользователей, а небольшой базовый пакет услуг провоцирует на дополнительные траты, что приносит значительную прибыль компании.

###  Подготовка данных

```python
users['reg_date'] = pd.to_datetime(users['reg_date'], format='%Y-%m-%dT%H:%M:%S')# обработка столбца reg_date
users['churn_date'] = pd.to_datetime(users['churn_date'], format='%Y-%m-%dT%H:%M:%S')# обработка столбца churn_date

calls['call_date'] = pd.to_datetime(calls['call_date'], format='%Y-%m-%dT%H:%M:%S')# обработка столбца call_date

messages['message_date'] = pd.to_datetime(messages['message_date'], format='%Y-%m-%dT%H:%M:%S')# обработка столбца message_date
sessions['session_date'] = pd.to_datetime(sessions['session_date'], format='%Y-%m-%dT%H:%M:%S')# обработка столбца session_date
```

```python
import numpy as np

calls['duration']= np.ceil(calls['duration']).astype(int)

# округление значений столбца duration с помощью np.ceil() 
# и приведение к типа данных int
```

Удаляю столбец Unnamed: 0 из датафрейма sessions. Столбец с таким названием возникает, когда данные сохраняют с указанием индекса (df.to_csv(..., index=column)). Он сейчас не понадобится.

```python
sessions = sessions.drop(['Unnamed: 0'], axis=1)
```

```python
calls['month'] = calls['call_date'].dt.month
```

```python
messages['month'] = messages['message_date'].dt.month
```

```python
sessions['month'] = sessions['session_date'].dt.month
```

```python
calls_per_month = calls.groupby(['user_id', 'month']).agg(calls=('duration', 'count'))
# подсчёт количества звонков для каждого пользователя по месяцам
calls_per_month.head(30)
```

```python
minutes_per_month = calls.groupby(['user_id', 'month']).agg(minutes=('duration', 'sum'))
# подсчёт израсходованных минут для каждого пользователя по месяцам
minutes_per_month.head(30)
```

```python
messages_per_month = messages.groupby(['user_id', 'month']).agg(messages=('message_date', 'count'))

# подсчёт количества отправленных сообщений для каждого пользователя по месяцам
messages_per_month.head(30)
```

```python
sessions_per_month = sessions.groupby(['user_id', 'month']).agg({'mb_used': 'sum'})
# подсчёт потраченных мегабайт для каждого пользователя по месяцам
sessions_per_month.head(30)
```

### Анализ данных и подсчёт выручки

```python
# churn_date - если значение пропущено, значит тариф ещё действует
users['churn_date'].count() / users['churn_date'].shape[0] * 100
# количество делится на общее количество пользователей (метод 'shape[0]') 
```

Расторгли договор 7.6% клиентов из датасета

```python
# Объединяю все посчитанные выше значения в один датафрейм user_behavior
user_behavior = calls_per_month\
    .merge(messages_per_month, left_index=True, right_index=True, how='outer')\
    .merge(sessions_per_month, left_index=True, right_index=True, how='outer')\
    .merge(minutes_per_month, left_index=True, right_index=True, how='outer')\
    .reset_index()\
    .merge(users, how='left', left_on='user_id', right_on='user_id')\

user_behavior.head()
```

```python
# Проверяю пропуски в таблице user_behavior после объединения:
user_behavior.isna().sum()
```

```python
# Заполним образовавшиеся пропуски в данных:
user_behavior['calls'] = user_behavior['calls'].fillna(0)
user_behavior['minutes'] = user_behavior['minutes'].fillna(0)
user_behavior['messages'] = user_behavior['messages'].fillna(0)
user_behavior['mb_used'] = user_behavior['mb_used'].fillna(0)
```

```python
# переименование столбца tariff_name на более простое tariff

tariffs = tariffs.rename(
    columns={
        'tariff_name': 'tariff'
    }
)
```

```python
# Присоединяем информацию о тарифах
user_behavior = user_behavior.merge(tariffs, on='tariff')
```

```python
user_behavior.head()
```

```python
# Считаю количество минут разговора, сообщений и мегабайт, превышающих включенные в тариф
user_behavior['paid_minutes'] = user_behavior['minutes'] - user_behavior['minutes_included']
user_behavior['paid_messages'] = user_behavior['messages'] - user_behavior['messages_included']
user_behavior['paid_mb'] = user_behavior['mb_used'] - user_behavior['mb_per_month_included']

for col in ['paid_messages', 'paid_minutes', 'paid_mb']:
    user_behavior.loc[user_behavior[col] < 0, col] = 0
```

```python
# Перевожу превышающие тариф мегабайты в гигабайты и сохраняем в столбец paid_gb
user_behavior['paid_gb'] = np.ceil(user_behavior['paid_mb'] / 1024).astype(int)
```

```python
# Считаю выручку за минуты разговора, сообщения и интернет
user_behavior['cost_minutes'] = user_behavior['paid_minutes'] * user_behavior['rub_per_minute']
user_behavior['cost_messages'] = user_behavior['paid_messages'] * user_behavior['rub_per_message']
user_behavior['cost_gb'] = user_behavior['paid_gb'] * user_behavior['rub_per_gb']
```

```python
# Считаю помесячную выручку с каждого пользователя, она будет храниться в столбец total_cost
user_behavior['total_cost'] = \
      user_behavior['rub_monthly_fee']\
    + user_behavior['cost_minutes']\
    + user_behavior['cost_messages']\
    + user_behavior['cost_gb']
```

```python
user_behavior
```

```python
# сохраняю статистические метрики для каждой пары месяц-тариф
# в одной таблице stats_df (среднее значение, стандартное отклонение, медиана)

stats_df = user_behavior.pivot_table(
            index=['month', 'tariff'],\
            values=['calls', 'minutes', 'messages', 'mb_used'],\
            aggfunc=['mean', 'std', 'median']\
).round(2).reset_index()

stats_df.columns=['month', 'tariff', 'calls_mean', 'sessions_mean', 'messages_mean', 'minutes_mean',
                                     'calls_std',  'sessions_std', 'messages_std', 'minutes_std', 
                                     'calls_median', 'sessions_median', 'messages_median',  'minutes_median']

stats_df.head(10)
```

## Проверка гипотез

Проверяю различаются ли средняя выручка пользователей тарифов «Ультра» и «Смарт»

```python
from scipy import stats as st
```

```python

ultra = user_behavior[user_behavior['tariff'] =='ultra']['total_cost']
smart = user_behavior[user_behavior['tariff'] =='smart']['total_cost']
alpha = 0.05 
# alpha = задаю значение уровня значимости

# провожу t-тест для сравнения средней выручки тарифов «Ультра» и «Смарт» 
results = st.ttest_ind(smart, ultra, equal_var = False)

print(results.pvalue)   

# P-значение - это вероятность того, что различие между средними значениями произошло случайно.

if results.pvalue < alpha:    
    print("Отвергаю нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Проверка гипотезы: пользователи из Москвы приносят больше выручки, чем пользователи из других городов;

H_0: Выручка (total_cost) пользователей из Москвы = выручка (total_cost) пользователей не из Москвы

H_1: Выручка (total_cost) пользователей из Москвы ≠ выручка (total_cost) пользователей не из Москвы

alpha = 0.05
