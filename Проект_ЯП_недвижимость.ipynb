```python
#напишем функцию категоризации по этажам, используя метод apply применимо к каждой строке датафрейма через axis = 1
def floor_category(row):
    floors_total = row['floors_total']
    floor = row['floor']
    if floor == 1:
        return 'первый'
    elif floor == floors_total:
        return 'последний'
    elif 1 < floor < floors_total:
        return 'другой'
    #категоризуем этажи с помощью функции  
data['floor_category'] = data.apply(floor_category, axis = 1)
```

### Линейный коэффициент корреляции r-Пирсона
Коэффициент корреляции Пирсона (r-Пирсона) применяется для исследования взаимосвязи двух переменных, измеренных в метрических шкалах на одной и той же выборке.
Коэффициент корреляции r-Пирсона характеризует существование линейной связи между двумя величинами. Если связь криволинейная то он не будет работать.

Чтобы приступать к расчетам коэффициента корреляции r-Пирсона необходимо выполнение следующих условий:

<li>  Исследуемые переменные X и Y должны быть распределены нормально.
<li>  Исследуемые переменные X и Y должны быть измерены в интервальной шкале или шкале отношений.
<li>  Количество значений в исследуемых переменных X и Y должно быть одинаковым.
    
<li> +1 - Полная положительная корреляция
<li> +0,8 - Сильная положительная корреляция
<li> +0.6 - Умеренная положительная корреляция
<li> 0 - никакой корреляции
<li> -0.6 - Умеренная отрицательная корреляция
<li> -0,8 - Сильная отрицательная корреляция
<li> -1 - Полная отрицательная корреляция
